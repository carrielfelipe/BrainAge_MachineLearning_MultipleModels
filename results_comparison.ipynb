```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Arial'

modelos = [
    "ElasticNet", "Lasso", "Ridge", "SVRlineal",
    "KernelRidge", "SVRkernel",
    "KNN", "BaggedKNN",
    "RandomForest", "XGBoost", "CatBoost",
    "MLP"
]

# Colores Earth tones por grupo de modelo
colores_algoritmos = {
    "ElasticNet": "#B5525C", "Lasso": "#C06C84", "Ridge": "#D98C8C", "SVRlineal": "#E8B4B8",
    "KernelRidge": "#4C6A92", "SVRkernel": "#6B8BA4",
    "KNN": "#6C9A8B", "BaggedKNN": "#88B0A0",
    "RandomForest": "#E1B07E", "XGBoost": "#C89F6D", "CatBoost": "#A77954",
    "MLP": "#3B3B3B"
}

base_dir = "results/"
modelos_dataframes = {}

for modelo in modelos:
    file_path = f"{modelo}/{base_dir}/df_CN_results_training.xlsx"
    modelos_dataframes[modelo] = pd.read_excel(file_path, engine='openpyxl')

for modelo in modelos_dataframes:
    modelos_dataframes[modelo] = modelos_dataframes[modelo].sort_values(by="ID_unique").reset_index(drop=True)

fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(9, 9))
axes = axes.flatten()

for i, modelo in enumerate(modelos):
    df = modelos_dataframes[modelo]
    ax = axes[i]

    ax.set_facecolor((1, 1, 1, 0.8))  # fondo blanco semitransparente

    for spine in ax.spines.values():
        spine.set_alpha(0.3)  # bordes tenues

    color = colores_algoritmos.get(modelo, "black")

    # Puntos y línea de regresión
    sns.scatterplot(x=df["y_labels"], y=df["y_pred_corrected"], alpha=0.99, color=color, edgecolor="none", ax=ax)
    sns.regplot(
        x=df["y_labels"], y=df["y_pred_corrected"], scatter=False,
        color='black', ci=None, ax=ax,
        line_kws={"linewidth": 1.5, "alpha": 0.7}
    )

    # Ajuste lineal
    coef = np.polyfit(df["y_labels"], df["y_pred_corrected"], deg=1)
    m = coef[0]
    y0 = coef[1]

    # Título del modelo
    ax.text(0.05, 0.95, modelo, transform=ax.transAxes,
            fontsize=10, verticalalignment='top',
            bbox=dict(facecolor='white', edgecolor='black', alpha=0.6))

    # Pendiente e intercepto
    ax.text(0.95, 0.05, f"m = {m:.2f}\ny$_0$ = {y0:.0f}", transform=ax.transAxes,
            fontsize=10, ha='right', va='bottom',
            bbox=dict(facecolor='white', edgecolor='black', alpha=0.6))

    ax.set_xticks([20, 90])
    ax.set_yticks([20, 90])
    ax.set_xticks(np.arange(0, 101, 10), minor=True)
    ax.set_yticks(np.arange(0, 101, 10), minor=True)

    ax.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.3)
    ax.tick_params(labelsize=10)

    if i % 4 != 0:
        ax.set_ylabel("")
        ax.set_yticklabels([])
    else:
        ax.set_ylabel("Brain Age", fontsize=10)

    if i < 8:
        ax.set_xlabel("")
        ax.set_xticklabels([])
    else:
        ax.set_xlabel("Chronological Age", fontsize=10)

    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)

plt.tight_layout()
plt.show()
```
