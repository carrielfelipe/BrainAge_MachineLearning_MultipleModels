```python
import pandas as pd
from collections import Counter

# Obtener nombres de columnas
columnas = X_CN_scaled.columns

# Extraer la región con lateralidad (últimos dos elementos del nombre de la columna)
regiones_lateralidad = ["_".join(col.split('_')[-2:]) for col in columnas]

# Contar ocurrencias por región con lateralidad
conteo_regiones = Counter(regiones_lateralidad)

# Convertir a DataFrame para mejor visualización
df_conteo = pd.DataFrame(conteo_regiones.items(), columns=['Región_Lateralidad', 'Cantidad']).sort_values(by='Cantidad', ascending=False)

print(df_conteo)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
fs = 250  # frecuencia de muestreo en Hz
t = np.linspace(0, 10, fs*10)  # 10 segundos de señal
freq_alpha = 10  # Hz

# Señal base + ruido
eeg_base = 50 * np.sin(2 * np.pi * freq_alpha * t)
noise = np.random.normal(0, 20, len(t))
eeg_signal = eeg_base + noise

# Gráfico limpio en negro
plt.figure(figsize=(8, 4))
plt.plot(t, eeg_signal, color='black', linewidth=1)

# Eliminar ejes, bordes, ticks, etc.
plt.axis('off')
plt.margins(0)
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal.windows import gaussian

# Parámetros
fs = 250  # Hz
dur = 2  # duración total en segundos
t = np.linspace(0, dur, int(fs * dur))
stim_time = 1.0  # instante del estímulo

# Crear la forma del ERP (N100 + P300)
n100_latency = 0.1  # Latencia del N100 (100 ms post-estímulo)
p300_latency = 0.3  # Latencia del P300 (300 ms post-estímulo)

# Duración de cada componente (en segundos)
erp_duration = 0.4
erp_samples = int(erp_duration * fs)

# Crear componentes de la onda
erp_wave = np.zeros_like(t)

# Componente N100 (pico negativo)
n100_center = int((stim_time + n100_latency) * fs)
n100_window = gaussian(erp_samples, std=fs * 0.02)  # gaussiana más estrecha
n100_window = -n100_window / np.max(n100_window) * 15  # pico negativo (-15 uV)
erp_wave[n100_center - erp_samples // 2:n100_center + erp_samples // 2] = n100_window

# Componente P300 (pico positivo)
p300_center = int((stim_time + p300_latency) * fs)
p300_window = gaussian(erp_samples, std=fs * 0.02)  # gaussiana similar
p300_window = p300_window / np.max(p300_window) * 20  # pico positivo (+20 uV)
erp_wave[p300_center - erp_samples // 2:p300_center + erp_samples // 2] += p300_window

# Señal total (solo ERP sin ruido)
eeg_signal = erp_wave

# Graficar limpio
plt.figure(figsize=(10, 4))
plt.plot(t, eeg_signal, color='black', linewidth=2)
#plt.axvline(x=stim_time, color='red', linestyle='--', linewidth=1, label="Estímulo")
plt.axis('off')
plt.xlim(0.5, 2)  # Limitar eje X entre 0 y 2 segundos
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
plt.show()
```
