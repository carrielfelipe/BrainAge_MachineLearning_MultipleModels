```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import pickle

file_path_CN = f'{path}Brain_Age/Data/CN_freq.xlsx'

df_CN = pd.read_excel(file_path_CN)


df_CN_filtrado = df_CN[(df_CN['Age'] >=20) & (df_CN['Age'] <= 90)].reset_index(drop=True)


X_CN = df_CN_filtrado[features]
y_CN = df_CN_filtrado["Age"]
ID_CN = df_CN_filtrado["ID_unique"]

#Scaler segun algoritmo
scaler = StandardScaler()
scaler.fit(X_CN)

X_CN_scaled = scaler.transform(X_CN)

X_CN_scaled = pd.DataFrame(X_CN_scaled, columns=X_CN.columns)

df_concatenado_CN = pd.concat([X_CN, y_CN, ID_CN], axis=1, ignore_index=False)
```

```python
import pandas as pd
from collections import Counter

# Obtener nombres de columnas
columnas = X_CN_scaled.columns

# Extraer la región de cada columna (suponiendo que la estructura es "_REGION_left" o "_REGION_right")
regiones = [col.split('_')[-2] for col in columnas]  # Toma el penúltimo elemento

# Contar ocurrencias por región
conteo_regiones = Counter(regiones)

# Convertir a DataFrame para mejor visualización
df_conteo = pd.DataFrame(conteo_regiones.items(), columns=['Región', 'Cantidad']).sort_values(by='Cantidad', ascending=False)

print(df_conteo)
```

```python
import pandas as pd
from collections import Counter

# Obtener nombres de columnas
columnas = X_CN_scaled.columns

# Extraer la región con lateralidad (últimos dos elementos del nombre de la columna)
regiones_lateralidad = ["_".join(col.split('_')[-2:]) for col in columnas]

# Contar ocurrencias por región con lateralidad
conteo_regiones = Counter(regiones_lateralidad)

# Convertir a DataFrame para mejor visualización
df_conteo = pd.DataFrame(conteo_regiones.items(), columns=['Región_Lateralidad', 'Cantidad']).sort_values(by='Cantidad', ascending=False)

print(df_conteo)
```

```python
import matplotlib.pyplot as plt

# Crear el gráfico de barras horizontal
plt.figure(figsize=(5, 5))
plt.barh(df_conteo['Región_Lateralidad'], df_conteo['Cantidad'], color='black')

# Personalizar etiquetas
plt.ylabel('Región con Lateralidad', fontsize=12)
plt.xlabel('Cantidad de Características', fontsize=12)
plt.title('Cantidad de Características por Región y Lateralidad', fontsize=14)
plt.gca().invert_yaxis()  # Invertir el eje Y para que el mayor esté arriba

# Mostrar valores al final de cada barra
for i, v in enumerate(df_conteo['Cantidad']):
    plt.text(v + 0.1, i, str(v), va='center', fontsize=10)

plt.show()
```
