```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Lista de modelos
modelos = [
    "ElasticNet", "Lasso", "Ridge", "SVRlineal",  # Lineales
    "KernelRidge", "SVRkernel",  # Kernel
    "KNN", "BaggedKNN",  # KNN
    "RandomForest", "XGBoost", "CatBoost",  # Árboles
    "MLP",  # Redes neuronales
    "Dummy"  # Baseline
]

# Directorio base
base_dir = "results/"

# Cargar el DataFrame principal con metadatos
#df_CN = pd.read_excel("ruta/a/tu/df_CN.xlsx")  # <- actualiza esta ruta si es necesario
metadata = df_CN[['ID_unique', 'Country_unique', 'Region']]

# Leer, mergear y guardar resultados
modelos_dataframes = {}

for modelo in modelos:
    parent_dir = os.path.dirname(os.getcwd())
    model_dir = os.path.join(parent_dir, modelo, base_dir)
    input_path = os.path.join(model_dir, 'df_CN_results_training.xlsx')
    output_path = os.path.join(model_dir, 'df_CN_results_training_with_country.xlsx')

    # Leer archivo original
    df_resultados = pd.read_excel(input_path, engine='openpyxl')

    # Merge con metadatos
    df_resultados = df_resultados.merge(metadata, on='ID_unique', how='left')

    # Guardar nuevo DataFrame
    df_resultados.to_excel(output_path, index=False)

    # Guardar en el diccionario también si lo necesitas en memoria
    modelos_dataframes[modelo] = df_resultados
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Lista de modelos
modelos = [
    "ElasticNet", "Lasso", "Ridge", "SVRlineal",
    "KernelRidge", "SVRkernel",
    "KNN", "BaggedKNN",
    "RandomForest", "XGBoost", "CatBoost",
    "MLP", "Dummy"
]

# Directorio base
base_dir = "results/"
parent_dir = os.path.dirname(os.getcwd())

# Estilo gráfico
sns.set(style="whitegrid")

for modelo in modelos:
    # Ruta del archivo Excel
    model_dir = os.path.join(parent_dir, modelo, base_dir)
    file_path = os.path.join(model_dir, 'df_CN_results_training_with_country.xlsx')
    
    # Leer datos
    df = pd.read_excel(file_path)

    # Verificar que exista la columna GAP_corrected y Region
    if 'GAP_corrected' in df.columns and 'Region' in df.columns:
        plt.figure(figsize=(10, 6))
        ax = sns.boxplot(x='Region', y='GAP_corrected', data=df, palette='Set2')
        ax.set_title(f'{modelo} - GAP_corrected por Región', fontsize=14)
        ax.set_ylabel('GAP_corrected')
        ax.set_xlabel('Región')
        plt.xticks(rotation=45)
        plt.tight_layout()

        # Mostrar gráfico
        plt.show()
    else:
        print(f"❌ {modelo}: columnas necesarias no encontradas.")
```
